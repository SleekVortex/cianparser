```python
import time
import pandas as pd

from cianparser.cianparser import CianParser
```

```python
parser = CianParser('Москва')
```

```python
n_pages = 53
rooms = (2, 3, 4, 5, 'studio')
```

```python
data = []
for room in rooms:
    for n_page in range(1, n_pages+1):
        flats = parser.get_flats(
            deal_type="sale", rooms=(room,),
            with_saving_csv=True,
            additional_settings={"start_page":n_page, "end_page":n_page},
            with_extra_data=True
        )
        data.extend(flats)
        time.sleep(180)
    time.sleep(240)
```

```python
df = pd.DataFrame(data)
df.to_csv('msk.csv', index=False)
df
```
