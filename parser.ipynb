```python
import time
import pandas as pd

from cianparser.cianparser import CianParser
```

```python
parser = CianParser('Москва')
```

```python
n_pages = 53
rooms = (2, 3, 4, 5, 'studio')
```

```python
time.sleep(600)
```

```python
for room in rooms:
    for n_page in range(3, n_pages+1):
        parser.get_flats(
            deal_type="sale", rooms=(room,),
            with_saving_csv=True,
            additional_settings={"start_page":n_page, "end_page":n_page},
            with_extra_data=True
        )
        time.sleep(180)
    time.sleep(240)
```

```python
import pandas as pd
import os

# Укажите путь к папке с CSV файлами
folder_path = os.getcwd()
# Укажите имя выходного файла
output_file = 'combined.csv'

# Список для хранения DataFrame
dataframes = []

# Проходим по всем файлам в указанной папке
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        # Читаем CSV файл и добавляем его в список
        df = pd.read_csv(file_path, sep=';', engine='python')
        dataframes.append(df)

# Объединяем все DataFrame в один
combined_df = pd.concat(dataframes, ignore_index=True)

# Сохраняем объединенный DataFrame в новый CSV файл
combined_df.to_csv(output_file, index=False)

print(f'Все CSV файлы из папки "{folder_path}" успешно объединены в "{output_file}".')
```
